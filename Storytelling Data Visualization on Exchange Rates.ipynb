#  Storytelling Visualization on Exchange Rates


### introduction to the Dataset

The dataset we'll use describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union.

If the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro (one euro has more value than one US dollar at this exchange rate).

Daria Chemkaeva put together the data set and made it available on [Kaggle](https://www.kaggle.com/datasets/lsind18/euro-exchange-daily-rates-19992020) — the data source is the European Central Bank. Note that the dataset gets regular updates — we downloaded it on January 2021.


```python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.head(5)
```

```python
exchange_rates.tail(5)
```

```python
exchange_rates.info()
```

### Data Cleaning

Our focus in this part of the project will be on the exchange rate between the euro and the American dollar.

Below, we do the following:

 - We rename the `[US dollar ]` and `Period\Unit`: columns to something easier to type — `US_dollar` and `Time`.
 - We change the `Time` column to a `datetime` data type.
 - We sort the values by `Time` in ascending order.
 - We reset the index (and drop the initial index).


```python
exchange_rates.rename(columns={'[US dollar ]':'US_dollar',
                               'Period\\Unit:':'Time'}, 
                     inplace=True)

exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
                              
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
print(euro_to_dollar)
```

```python
# see if anything wrong
euro_to_dollar['US_dollar'].value_counts()
```

```python
# drop all the rows where the - chacarater appears in the Us_dollar column.
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar']!='-']
# chance the data type to float
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype('float')
euro_to_dollar.info()
```

# Rolling Mean plot

 we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar['Time'], 
        euro_to_dollar['US_dollar'].rolling(30).mean())
plt.show()
```

If we look at the line's shape,have meaning: they are the visual representation of the daily variation in the exchange rate. The rate goes up and down, up and down again, day to day. 

we are going to calculate the rolling mean with a moving window of 30 days.

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(7).mean()
print(euro_to_dollar['rolling_mean'] )
```

```python
%matplotlib inline
plt.plot(euro_to_dollar['rolling_mean'])
plt.show()
```

```python
%matplotlib inline
plt.figure(figsize=(10,12))
plt.subplot(3, 2, 1)
plt.plot(euro_to_dollar['Time'],
        euro_to_dollar['US_dollar'])
plt.title('original values',weight='bold')

plt.subplot(3, 2, 2)
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'].rolling(7).mean())
plt.title('rolling_mean_7', weight='bold')
plt.subplot(3, 2, 3)
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'].rolling(30).mean())
plt.title('rolling_mean_30',weight='bold')
plt.subplot(3, 2, 4)
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'].rolling(50).mean())
plt.title('rolling_mean_50',weight='bold')
plt.subplot(3, 2, 5)
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'].rolling(100).mean())
plt.title('rolling_mean_100',weight='bold')
plt.subplot(3, 2, 6)
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'].rolling(365).mean()
)
plt.title('rolling_mean_365',weight='bold')
plt.tight_layout()  #Auto-adjusts the padding between subplots
plt.show()

```

Above, we see how our line plot changes visually as we increase the rolling window.  this show us the long-term trends.

### Coming up with an Idea
Our goal is to create a storytelling data visualization using the data we have.

Here are a few story ideas for our data:

- We show how the euro-dollar rate has changed during the coronavirus pandemic. We can show the 2020 data and the 2016-2019 data as a baseline. We can use a line plot.
- We show how the euro-dollar rate changed during the 2007-2008 financial crisis. We can also show the data for 2006 and 2009 for comparison. We can use a line plot.
- We show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)). We can use a line plot.

## Storytelling Data Visualization examples 

### Financial Crisis 2008

```python
financial_crisis = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >=2006
                                         ) & (euro_to_dollar['Time'].dt.year <=2009)]
financial_crisis_7_8 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >=2007)
                                            & (euro_to_dollar['Time'].dt.year <=2008)]
print(financial_crisis_7_8 )
```

```python
# adding the FiveThirtyEight Style
import matplotlib.style as style
style.use('fivethirtyeight')

#adding a plot
fig, ax=plt.subplots(figsize=(8,3))
ax.plot(financial_crisis['Time'],
        financial_crisis['rolling_mean'],
        linewidth= 1, color='#665628')

#adding hightlighting the 2007-2008 perios
ax.plot(financial_crisis_7_8['Time'],
        financial_crisis_7_8['rolling_mean'],
        linewidth=3, color='#283866')

#adding hightlighting the peak of crisis
ax.axvspan(xmin=733112.0, xmax=733302.0, ymin=0.09,
           alpha=0.3, color='black')

#adding separated ticklabel

ax.set_xticklabels([])
ax.set_yticklabels([])

x = 732272.0
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, 1.13, year, alpha = 0.8, fontsize=12)
    x+=365
    
y= 1.193
for rates in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(732172.0, y, rates, alpha=0.8, fontsize=12)
    y+=0.1
    
#adding title and a subtitle
ax.text(732172.0, 1.67,"Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",
       weight='bold')
ax.text(732172.0, 1.63, 'Euro-USD exchange rates between 2006 and 2010',
        size=12)

# adding a signature
ax.text(732272.0, 1.07,'©DATAQUEST' +' '*88 + 'Source: European Central Bank',
color='#f0f0f0', backgroundcolor ='#4d4d4d', size= 10)

#adding some transparency to the grid
ax.grid(alpha=0.5)
plt.show()
```

###  The last Three US Presidents

```python
bush_obama_trump = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year>=2001)
                                        & (euro_to_dollar['Time'].dt.year < 2021)]

george_bush = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year < 2009)]
                                    
barack_obama = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year>=2009) & (bush_obama_trump['Time'].dt.year < 2017)]
donal_trump = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >=2017) & (bush_obama_trump['Time'].dt.year < 2021)]
```

```python
#adding the fivethirtyeight style
import matplotlib.style as style
style.use('fivethirtyeight')

plt.figure(figsize = (12,6))
ax1=plt.subplot(2, 3, 1)
ax2=plt.subplot(2, 3, 2)
ax3=plt.subplot(2, 3, 3)
ax4=plt.subplot(2, 1, 2)
axes =[ax1, ax2, ax3, ax4]

#changes all to subplots
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0','1.2', '1.4','1.4'],
                      alpha=0.3)
    ax.grid(alpha=0.5)
    
#adding ax1=bush
ax1.plot(george_bush['Time'],
         george_bush['rolling_mean'],
        color='#16537e')
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '',
                     '2007', '', '2009'], alpha=0.3)
ax1.text(731516.0, 1.92, 'BUSH', fontsize= 18, color='#16537e',
         weight='bold')

ax1.text(731216.0, 1.8, '(2001-2009)', weight='bold',
         color='#16537e')

#adding ax2 = barack obama
ax2.plot(barack_obama['Time'],
         barack_obama['rolling_mean'],
         color='#783f04')

ax2.set_xticklabels(['','2009','','2011','','2013','','2015','','2017'],
                   alpha=0.3)
ax2.text(734288.0, 1.92,'OBAMA', fontsize= 18, color='#783f04',
        weight='bold')

ax2.text(734138.0, 1.8, '(2008-2017)', weight='bold', color='#783f04')

#adding ax3 = trump
ax3.plot(donal_trump['Time'],
         donal_trump['rolling_mean'],
         color='#8fce00')
ax3.set_xticklabels(['2017','','2018','','2019','','2020','','2021'],
                    alpha=0.3)
ax3.text(736855.0, 1.9, 'TRUMP', fontsize=18, color='#8fce00')
ax3.text(736745.0, 1.8, '(2017-2021)', weight='bold',color='#8fce00')

#adding ax4 = all
ax4.plot(george_bush['Time'],
         george_bush['rolling_mean'],
        color='#16537e')
ax4.plot(barack_obama['Time'],
         barack_obama['rolling_mean'],
         color='#783f04')
ax4.plot(donal_trump['Time'],
         donal_trump['rolling_mean'],
         color='#8fce00')
ax4.grid(alpha=0.3)
ax4.set_xticks([])

#adding title and subtitle
ax1.text(730016.0, 2.35, 'EURO-USD rate averaged 1.22 under the last three US presidents',
         fontsize=20, weight='bold')
ax1.text(730016.0, 2.14, '''EURO-USD exchange rates under George W. Bush (2001 - 2009),
Barack Obama (2009-2017), and Donald Trump (2017-2021)''',
        fontsize=16)

#adding asignature
ax4.text(729916.0, 0.65, '©DATAQUEST' + ' '*103 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=14)
```
